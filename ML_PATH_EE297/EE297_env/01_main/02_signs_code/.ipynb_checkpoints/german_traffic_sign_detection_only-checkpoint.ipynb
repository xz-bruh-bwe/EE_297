```python
def rescale_bbox(x1, y1, x2, y2, orig_w, orig_h, target_w=224, target_h=224):
    """
    Rescales and normalizes bounding box from original image size to target size.
    Returns: normalized [x_center, y_center, width, height]
    """
    scale_x = target_w / orig_w
    scale_y = target_h / orig_h

    # Rescale coordinates
    x1_scaled = x1 * scale_x
    y1_scaled = y1 * scale_y
    x2_scaled = x2 * scale_x
    y2_scaled = y2 * scale_y

    # Compute center + width/height
    box_w = x2_scaled - x1_scaled
    box_h = y2_scaled - y1_scaled
    x_center = (x1_scaled + x2_scaled) / 2
    y_center = (y1_scaled + y2_scaled) / 2

    # Normalize
    x_center /= target_w
    y_center /= target_h
    box_w /= target_w
    box_h /= target_h

    return x_center, y_center, box_w, box_h
```

### Bounding Boxes Check

```python
def rescale_bbox(x1, y1, x2, y2, orig_w, orig_h, target_w=224, target_h=224):
    """
    Rescales and normalizes bounding box from original image size to target size (e.g., 224x224).
    Returns: x_center, y_center, width, height (all normalized to [0, 1])
    """
    scale_x = target_w / orig_w
    scale_y = target_h / orig_h

    x1_scaled = x1 * scale_x
    y1_scaled = y1 * scale_y
    x2_scaled = x2 * scale_x
    y2_scaled = y2 * scale_y

    w = x2_scaled - x1_scaled
    h = y2_scaled - y1_scaled
    x_center = (x1_scaled + x2_scaled) / 2
    y_center = (y1_scaled + y2_scaled) / 2

    return x_center / target_w, y_center / target_h, w / target_w, h / target_h
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

def visualize_bbox(img_path, bbox, image_size=(224, 224)):
    """
    Draws a bounding box on the image.
    `bbox`: (x_center, y_center, width, height) in normalized format
    """
    image = Image.open(img_path).convert("RGB").resize(image_size)
    fig, ax = plt.subplots(1)
    ax.imshow(image)

    x_c, y_c, w, h = bbox
    x = (x_c - w / 2) * image_size[0]
    y = (y_c - h / 2) * image_size[1]
    w = w * image_size[0]
    h = h * image_size[1]

    rect = patches.Rectangle((x, y), w, h, linewidth=2, edgecolor='r', facecolor='none')
    ax.add_patch(rect)
    plt.title("Bounding Box Overlay")
    plt.axis("off")
    plt.show()
```

```python
class SignDetectionDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.df = dataframe.reset_index(drop=True)
        self.transform = transform
        self.target_size = (224, 224)

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.loc[idx]
        img_path = row["full_path"]  # make sure this is absolute
        class_id = int(row["LabelIndex"])
    
        image = Image.open(img_path).convert("RGB")
        image = image.resize(self.target_size)
    
        x_center, y_center, box_w, box_h = rescale_bbox(
            row["Roi.X1"], row["Roi.Y1"], row["Roi.X2"], row["Roi.Y2"],
            row["Width"], row["Height"],
            *self.target_size
        )
    
        if self.transform:
            image = self.transform(image)
    
        return image, {
            "bbox": torch.tensor([x_center, y_center, box_w, box_h], dtype=torch.float32),
            "label": torch.tensor(class_id, dtype=torch.long)
        }
```

```python
def custom_collate_fn(batch):
    images = []
    bboxes = []
    labels = []

    for img, target in batch:
        images.append(img)
        bboxes.append(target["bbox"])
        labels.append(target["label"])

    return torch.stack(images), torch.stack(bboxes), torch.stack(labels)
```

```python
class MultiTaskNet(nn.Module):
    def __init__(self, num_classes):
        super(MultiTaskNet, self).__init__()

        resnet = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
        self.encoder = nn.Sequential(*list(resnet.children())[:-2])  # (B, 512, 7, 7)

        # Detection head
        self.detection_head = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Flatten(),
            nn.Linear(512, 128),
            nn.ReLU(),
        )
        self.bbox_regressor = nn.Linear(128, 4)       # x_center, y_center, width, height
        self.classifier = nn.Linear(128, num_classes) # class logits

        # Segmentation head (unchanged)
        self.segmentation_head = nn.Sequential(
            nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),

            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),

            nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),

            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),

            nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),

            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),

            nn.ConvTranspose2d(64, 32, kernel_size=2, stride=2),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),

            nn.Conv2d(32, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),

            nn.Conv2d(32, 1, kernel_size=1),
            nn.Upsample(size=(360, 640), mode='bilinear', align_corners=False)
        )

    def forward(self, x):
        features = self.encoder(x)

        # Detection head
        shared_det_features = self.detection_head(features)
        bbox_out = self.bbox_regressor(shared_det_features)
        class_out = self.classifier(shared_det_features)

        # Segmentation head
        seg_out = self.segmentation_head(features)

        return bbox_out, class_out, seg_out
```

### TRAINING

```python
import torch
import torch.nn as nn
from tqdm import tqdm

# === Device Setup ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = MultiTaskNet(num_classes=10).to(device)

# === Freeze Segmentation Head Only ===
for param in model.segmentation_head.parameters():
    param.requires_grad = False

# Unfreeze Detection Parts
for param in model.encoder.parameters():
    param.requires_grad = True
for param in model.detection_head.parameters():
    param.requires_grad = True
for param in model.bbox_regressor.parameters():
    param.requires_grad = True
for param in model.classifier.parameters():
    param.requires_grad = True

# === Loss Functions (Scalar Reduction) ===
criterion_bbox = nn.SmoothL1Loss(reduction='mean')
criterion_cls = nn.CrossEntropyLoss(reduction='mean')

# === Optimizer ===
optimizer = torch.optim.Adam([
    {'params': model.encoder.parameters(), 'lr': 1e-4},
    {'params': model.detection_head.parameters(), 'lr': 1e-3},
    {'params': model.bbox_regressor.parameters(), 'lr': 1e-3},
    {'params': model.classifier.parameters(), 'lr': 1e-3}
])

# === Training Loop ===
EPOCHS = 20

for epoch in range(EPOCHS):
    model.train()
    train_loss = 0.0
    print(f"\nEpoch {epoch+1}/{EPOCHS}")

    for images, bbox_targets, class_targets in tqdm(train_loader, desc="Train"):
        images = images.to(device)
        bbox_targets = bbox_targets.to(device)
        class_targets = class_targets.to(device)

        optimizer.zero_grad()

        # Forward
        bbox_out, class_out, _ = model(images)

        # Losses
        loss_bbox = criterion_bbox(bbox_out, bbox_targets)
        loss_cls = criterion_cls(class_out, class_targets)
        loss = loss_bbox + loss_cls

        # ðŸ§  Make sure it's scalar
        assert loss.dim() == 0, f"Expected scalar loss, got shape: {loss.shape}"

        # Backward
        loss.backward()
        optimizer.step()

        train_loss += loss.item()

    avg_train_loss = train_loss / len(train_loader)
    print(f"Train Loss: {avg_train_loss:.4f}")

    # === Validation ===
    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for images, bbox_targets, class_targets in tqdm(val_loader, desc="Val"):
            images = images.to(device)
            bbox_targets = bbox_targets.to(device)
            class_targets = class_targets.to(device)

            bbox_out, class_out, _ = model(images)

            loss_bbox = criterion_bbox(bbox_out, bbox_targets)
            loss_cls = criterion_cls(class_out, class_targets)
            loss = loss_bbox + loss_cls

            val_loss += loss.item()

    avg_val_loss = val_loss / len(val_loader)
    print(f"Val Loss: {avg_val_loss:.4f}")
```

```python
def denormalize_bbox(bbox, image_size=(224, 224)):
    """
    Convert normalized bbox (x_center, y_center, w, h in [0,1]) to pixel coordinates:
    (top-left x, top-left y, width, height) in pixels.
    """
    x_c, y_c, w, h = bbox
    W, H = image_size

    x = (x_c - w / 2) * W
    y = (y_c - h / 2) * H
    w = w * W
    h = h * H

    return x, y, w, h
```

```python
def show_predictions(model, dataloader, device, class_names, num_samples=5, save_csv=False):
    import csv
    import matplotlib.pyplot as plt
    import matplotlib.patches as patches

    model.eval()
    images, bbox_targets, class_targets = next(iter(dataloader))

    images = images.to(device)
    bbox_targets = bbox_targets.to(device)
    class_targets = class_targets.to(device)

    with torch.no_grad():
        bbox_out, class_out, _ = model(images)
        pred_classes = class_out.argmax(dim=1)

    images = images.cpu()
    predictions = []

    fig, axs = plt.subplots(1, num_samples, figsize=(18, 5))

    for i in range(num_samples):
        img = images[i].permute(1, 2, 0)
        img = img * torch.tensor([0.229, 0.224, 0.225]) + torch.tensor([0.485, 0.456, 0.406])
        img = img.clip(0, 1)

        ax = axs[i]
        ax.imshow(img)

        true_box = denormalize_bbox(bbox_targets[i].cpu().tolist())
        pred_box = denormalize_bbox(bbox_out[i].cpu().tolist())

        rect_true = patches.Rectangle((true_box[0], true_box[1]), true_box[2], true_box[3],
                                      linewidth=2, edgecolor='green', facecolor='none', label='GT')
        rect_pred = patches.Rectangle((pred_box[0], pred_box[1]), pred_box[2], pred_box[3],
                                      linewidth=2, edgecolor='red', facecolor='none', label='Pred')

        ax.add_patch(rect_true)
        ax.add_patch(rect_pred)

        true_label = class_targets[i].item()
        pred_label = pred_classes[i].item()

        ax.set_title(f"T:{class_names[true_label]} / P:{class_names[pred_label]}", fontsize=10)
        ax.axis('off')

        # Save prediction info
        predictions.append({
            "True Label": class_names[true_label],
            "Predicted Label": class_names[pred_label],
            "Predicted BBox": pred_box
        })

    plt.tight_layout()
    plt.show()
```

```python
from PIL import Image
import torch
import torchvision.transforms as transforms

# === Load and Preprocess Image ===
def load_image(image_path):
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225])
    ])

    image = Image.open(image_path).convert("RGB")
    image_tensor = transform(image).unsqueeze(0)  # Shape: (1, 3, 224, 224)
    return image_tensor, image

def run_inference_on_image(model, image_path, class_names=None, device='cuda'):
    model.eval()
    image_tensor, original_image = load_image(image_path)
    image_tensor = image_tensor.to(device)

    with torch.no_grad():
        bbox_out, class_out, _ = model(image_tensor)
        predicted_class = class_out.argmax(dim=1).item()
        predicted_bbox = bbox_out.squeeze().cpu().tolist()  # (4,)

    class_name = class_names[predicted_class] if class_names else str(predicted_class)

    print(f"ðŸ§  Predicted Class: {class_name} (index: {predicted_class})")
    print(f"ðŸ“¦ Predicted BBox (normalized [x_center, y_center, w, h]): {predicted_bbox}")

    return original_image, image_tensor, predicted_bbox, predicted_class
```

```python
image_path = "test_2.jpg"
original_img, image_tensor, pred_bbox, pred_class = run_inference_on_image(
    model, image_path, class_names=gtsrb_legend, device=device
)

draw_predicted_box_on_resized(image_tensor, pred_bbox, label=gtsrb_legend[pred_class])
```
